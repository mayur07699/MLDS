```python
num = ['team1_runs', 'team1_wickets', 'team2_runs', 'team2_wickets',
       'win_by_runs', 'win_by_wickets', 'overs_limit',
       'team1_wickets_remaining', 'team2_run_rate', 'total_wickets_fell',
       'run_margin_dup', 'wicket_margin_dup', 'team2_runs_x2']
categories_cols = ["team1", "team2", "venue", "city", "toss_winner",
"toss_decision",
"season", "player_of_match"]
X = pd.get_dummies(
    df[num + categories_cols],
    columns=categories_cols,
    drop_first=True
)
Y = df['team1_runs']
print("Final feature matrix shape:", X.shape)
print("Number of rows (matches):", len(df))
```

```python
X_train,X_test,Y_train,Y_test =train_test_split(X,Y,test_size= 0.25,random_state=42)
scalar = StandardScaler()
X_train_s = scalar.fit_transform(X_train)
X_test_s = scalar.fit_transform(X_test)
print("Train shape",X_train_s.shape)
print("Train shape:", X_train_s.shape)
print("Test shape :", X_test_s.shape)

```

```python
lr = LinearRegression()
lr.fit(X_train_s,Y_train)
lr_train_r2 = r2_score(Y_train, lr.predict(X_train_s))
lr_test_r2 = r2_score(Y_test,lr.predict(X_test_s))
lr_gap = lr_train_r2 -lr_test_r2
print(lr_gap)
```

```python
#Ridgee
lambdas = np.array([0.1, 1, 5, 10, 20, 100, 200, 300, 500, 800, 1000, 1500, 2000, 3000])

train_scores = []
test_scores = []

for alpha in lambdas:
    ridge = Ridge(alpha=alpha)
    ridge.fit(X_train_s, Y_train)

    train_pred = ridge.predict(X_train_s)
    test_pred = ridge.predict(X_test_s)

    train_scores.append(r2_score(Y_train, train_pred))
    test_scores.append(r2_score(Y_test, test_pred))

result = pd.DataFrame({
    "lambda": lambdas,
    "train_r2": train_scores,
    "test_r2": test_scores
})

print(f"Linear Regression -> Train R2: {lr_train_r2:.4f}")
print(f"Linear Regression -> Test  R2: {lr_test_r2:.4f}")
print(f"Train - Test gap (overfitting signal): {lr_gap:.4f}")
```

```python
from sklearn.linear_model import RidgeCV
from sklearn.metrics import r2_score
import numpy as np

# Candidate alpha values
alphas = np.array([0.1, 1, 5, 10, 20, 100, 200, 300, 500, 800, 1000, 1500, 2000, 3000])

# Ridge Regression with Cross-Validation
ridge_cv = RidgeCV(alphas=alphas, cv=5)
ridge_cv.fit(X_train_s, Y_train)

# Best alpha selected by the model
print("Best lambda (alpha):", ridge_cv.alpha_)

# Predictions
train_pred = ridge_cv.predict(X_train_s)
test_pred = ridge_cv.predict(X_test_s)

# R² scores
print("Train R²:", r2_score(Y_train, train_pred))
print("Test R²:", r2_score(Y_test, test_pred))
```
