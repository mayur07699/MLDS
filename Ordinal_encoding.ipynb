```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv("/content/IPLPlayerStat-1.csv")
```

```python
df.head()
```

```python
def player_form(avg):
  if avg <= 10:
    return "poor"
  elif avg <= 15:
    return "average"
  elif avg <= 20:
    return "good"
  elif avg <= 25:
    return "very good"
  else:
    return "excellent"
df['Player_Form'] = df['batting_avg'].apply(player_form)
df[['player','batting_avg','Player_Form']].head()
```

```python
df
```

```python
df['Player_Form'].unique()
```

```python
encoder = OrdinalEncoder(
    categories = [['poor', 'average', 'very good', 'good', 'excellent']]
)
df['Player_Form_encoded']=encoder.fit_transform(df[['Player_Form']]).astype(int)
df[['player','batting_avg','Player_Form_encoded']].head()
```
