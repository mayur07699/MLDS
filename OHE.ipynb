```python
df_encoded = pd.get_dummies(
    df,
    columns=['team1','team2','toss_decision'],
    dtype=int
)
pd.set_option('display.max_columns', None)
df_encoded.head(5)
```

```python
print(df_encoded.columns)
```

```python
df_encoded2 = pd.get_dummies(
    df,
    columns=['team1','team2','toss_decision'],
    dtype=int,
    drop_first=True
)
pd.set_option('display.max_columns', None)
df_encoded2.head(5)
```
