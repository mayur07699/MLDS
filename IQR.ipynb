```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('/content/IPLPlayerStat-1.csv')
```

```python
Q1 = df['bowling_strike_rate'].quantile(0.25)
Q3 = df['bowling_strike_rate'].quantile(0.75)
median = df['bowling_strike_rate'].quantile(0.5)

iqr = Q3 - Q1

lower = Q1 - 1.5 * iqr
upper = Q3 + 1.5 * iqr

print("Q1     :", Q1)
print("Median :", median)
print("Q3     :", Q3)
print("IQR    :", iqr)
print("Lower  :", lower)
print("Upper  :", upper)
```

```python
outlier1 = (df['bowling_strike_rate'] < lower) | (df['bowling_strike_rate'] > upper)
```
